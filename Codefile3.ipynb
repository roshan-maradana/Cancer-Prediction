```python

# Combine first 38 rows from both DataFrames
combined_df = pd.concat([stromal_df.head(23), tumor_df.head(23)])
combined_df
```

```python
target_values = [0]*23 + [1]*23  # 0 is immune cell 1 is tumor cell
combined_df['target'] = target_values
combined_df
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
X = combined_df.drop(columns=['target'])  # all columns except 'target'
y = combined_df['target']                 # target column

# Split data into training and testing sets

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,         # 20% data for testing # By default, train_test_split does shuffle the data before splitting.
    random_state=42,       # for reproducibility
    stratify=y             # important if target classes are imbalanced (guarantees that both train and test will have the same proportion of class 0 and class 1!)
)

from sklearn.preprocessing import StandardScaler

# Convert all column names to strings before scaling
X_train.columns = X_train.columns.astype(str)
X_test.columns = X_test.columns.astype(str)

# Center only (no scaling)
scaler = StandardScaler(with_mean=True, with_std=False)
X_train_centered = scaler.fit_transform(X_train)
X_test_centered = scaler.transform(X_test)  # Use training set's mean

from sklearn.decomposition import PCA

pca = PCA(n_components=2) # Keep components explaining 95% variance
X_train_pca = pca.fit_transform(X_train_centered)
X_test_pca = pca.transform(X_test_centered)

print(pca.explained_variance_ratio_)
print(pca.explained_variance_ratio_.sum())

import matplotlib.pyplot as plt
plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=y_train, cmap='coolwarm', edgecolor='k', alpha=0.7)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA Projection')
plt.show()

loadings = pca.components_

# 2. Select top components based on explained variance
num_components_to_consider = 2  # Adjust as needed
top_components = loadings[:num_components_to_consider]

# 3. Select features with high loadings for each component
num_features_to_select = 500  # Adjust as needed
important_feature_indices = []
for component in top_components:
    top_feature_indices = np.argsort(np.abs(component))[-num_features_to_select:]
    important_feature_indices.extend(top_feature_indices)

# 4. Get feature names and create a new DataFrame
important_features = X_train.columns[important_feature_indices]  # Use X_train.columns
important_features = list(set(important_features))  # Remove duplicates
X_test_selected = X_test[important_features]

# Now you have X_test_selected with the selected features

important_features_df = pd.DataFrame(important_features, columns=['Important_Features'])
important_features_df.to_csv('fimportant features from tum and b cell.csv', index=False) # Save to CSV
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Train RandomForest on the PCA-transformed training data
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train_pca, y_train)

# Predict on the PCA-transformed test data
y_pred = rf.predict(X_test_pca)

# Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np

# Define classifier
rf = RandomForestClassifier(random_state=42)

# Define Stratified K-Fold
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)  # 5 folds

# Perform Cross-Validation
scores = cross_val_score(rf, X, y, cv=kfold, scoring='accuracy')  # You can also use 'f1', 'roc_auc' etc.

print("Cross-Validation Accuracy Scores:", scores)
print("Mean Accuracy:", np.mean(scores))
print("Standard Deviation:", np.std(scores))

# To get confusion matrix and classification report from cross-validation:
# Need to do manual cross-validation (because cross_val_score only gives scores)

from sklearn.model_selection import cross_val_predict

y_pred = cross_val_predict(rf, X, y, cv=kfold)

print("\nConfusion Matrix:\n", confusion_matrix(y, y_pred))
print("\nClassification Report:\n", classification_report(y, y_pred))
```
